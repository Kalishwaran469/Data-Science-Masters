```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
titanic=sns.load_dataset('titanic')

df=pd.DataFrame(titanic)

def normalize(column):
    return (column-column.min())/(column.max()-column.min())

df['normalized_age']=normalize(df['age'])
df['left_skewed']=df['normalized_age']*0.5
df['right_skewed']=df['normalized_age']*2

plt.figure(figsize=(10,5))
sns.boxplot(data=df[['normalized_age','left_skewed','right_skewed']])
plt.ylim(0,1)
plt.xlabel('column')
plt.ylabel('value')
plt.title('box_plot')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
titanic=sns.load_dataset('titanic')

df=pd.DataFrame(titanic)

def normalize(column):
    return (column-column.min())/(column.max()-column.min())

df['normalized_age']=normalize(df['age'])

df['left_skewed']=df['normalized_age']-0.2
df.loc[df['left_skewed']<-0.2,'left_skewed']-=0.5

df['right_skewed']=df['normalized_age']+0.2
df.loc[df['right_skewed']>1.2,'right_skewed']+=0.5

plt.figure(figsize=(10,5))
sns.boxplot(data=df[['normalized_age','left_skewed','right_skewed']])
plt.ylim(-0.5,1.5)
plt.xlabel('column')
plt.ylabel('value')
plt.title('box_plot')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
iris=sns.load_dataset('iris')
import numpy as np

df=pd.DataFrame(iris)

def normalize(column):
    return (column-column.min())/(column.max()-column.min())

df['normalized_sepal_width']=normalize(df['sepal_width'])

left_skewed_data=np.random.normal(loc=-0.2,scale=0.05,size=len(df))
df['left_skewed']=df['normalized_sepal_width']-left_skewed_data

right_skewed_data=np.random.normal(loc=0.2,scale=0.05,size=len(df))
df['right_skewed']=df['normalized_sepal_width']+right_skewed_data

plt.figure(figsize=(10,5))
sns.boxplot(data=df[['normalized_sepal_width','left_skewed','right_skewed']])
plt.ylim(0,1)
plt.xlabel('column')
plt.ylabel('value')
plt.title('box_plot')
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
iris=sns.load_dataset('iris')
import numpy as np

df=pd.DataFrame(iris)

# def normalize(column):
 #    return (column-column.min())/(column.max()-column.min())

normalized_data=(df['sepal_width']-df['sepal_width'].mean())/df['sepal_width'].std()

plt.hist(normalized_data,bins=30,edgecolor='black')
plt.xlabel('normalized_sepal_width')
plt.ylabel('frequency')
plt.title('histogram of normalized sepal width')
plt.show()

high_kurtosis_data=(df['sepal_width']-df['sepal_width'].mean())*5
                    
plt.hist(high_kurtosis_data,bins=30,edgecolor='black')
plt.xlabel('high_kurtosis_sepal_width')
plt.ylabel('frequency')
plt.title('histogram of high kurtosis sepal width')
plt.show()
                    
low_kurtosis_data=(df['sepal_width']-df['sepal_width'].mean())/5
                    
plt.hist(low_kurtosis_data,bins=30,edgecolor='black')
plt.xlabel('low_kurtosis_sepal_width')
plt.ylabel('frequency')
plt.title('histogram of low kurtosis sepal width')
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
iris=sns.load_dataset('iris')
import numpy as np

df=pd.DataFrame(iris)

# def normalize(column):
 #    return (column-column.min())/(column.max()-column.min())

sepal_length=df['sepal_length']

plt.hist(sepal_length,bins=30,edgecolor='black')
plt.xlabel('original_sepal_length')
plt.ylabel('frequency')
plt.title('histogram of original sepal length')
plt.show()

laptokrytic_data=sepal_length+2
                    
plt.hist(laptokrytic_data,bins=30,edgecolor='black')
plt.xlabel('laptokrytic_sepal_length')
plt.ylabel('frequency')
plt.title('histogram of laptokrytic sepal length')
plt.show()
                    
platykurtic_data=sepal_length-2
                    
plt.hist(platykurtic_data,bins=30,edgecolor='black')
plt.xlabel('platykurtic_sepal_length')
plt.ylabel('frequency')
plt.title('histogram of platykurtic sepal length')
plt.show()




```
