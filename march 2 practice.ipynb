```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(30)
y=np.random.rand(30)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(3,6))
plt.scatter(x,y,c='#FF3355',alpha=0.5,marker='x')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my graph of x vs y')
plt.grid()
```

```python
x=np.linspace(1,10,50)
y=np.cos(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y,'--r')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my cosine graph presentation')
plt.grid()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
```

```python
fig=plt.figure()
jk=fig.add_subplot(projection='3d')
jk.scatter(x,y,z,c='r',alpha=.8)
plt.xlabel('this is my x value')
plt.ylabel('this is my y value')
plt.title('this is my 3 dimensional presentation')
plt.show()
```
