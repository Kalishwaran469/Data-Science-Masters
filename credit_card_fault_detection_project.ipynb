```python
x=df.drop(labels=['default payment next month'],axis=1)
```

```python
y=df['default payment next month']
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()


```

```python
x_train_scaled=scaler.fit_transform(x_train)
```

```python
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(x_train_scaled,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=clf.predict(x_test_scaled)
accuracy_score(y_test,y_pred)*100
```

```python
param_grid={"var_smoothing":[0.1, 0.001, 0.5,0.05,0.01,1e-8,1e-7,1e-6,1e-10,1e-11]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gs=GridSearchCV(clf,param_grid=param_grid,cv=5,verbose=3)
```

```python
gs.fit(x_train_scaled,y_train)
```

```python
gs.best_params_
```

```python
clff=GaussianNB(var_smoothing = 0.5)
clff.fit(x_train_scaled,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred1=clff.predict(x_test_scaled)
accuracy_score(y_test,y_pred1)*100
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train_scaled,y_train)
y_pred2=rf.predict(x_test_scaled)
accuracy_score(y_test,y_pred2)*100
```

```python
param_grid_rf = { "n_estimators": [50,100,130,200],
                    "max_depth": range(3,11,1),
                    "random_state":[0,50,100],
                     "criterion":['gini',"entropy"]
                     }
```

```python
gs1=GridSearchCV(rf,param_grid=param_grid_rf,cv=5,verbose=3)
```

```python
gs1.fit(x_train_scaled,y_train)
```

```python
gs1.best_params_
```

```python
rff=RandomForestClassifier(criterion='entropy',
 max_depth= 8,
 n_estimators= 500,
 random_state= 0)
rff.fit(x_train_scaled,y_train)
y_pred3=rff.predict(x_test_scaled)
accuracy_score(y_test,y_pred3)*100
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train_scaled,y_train)
y_pred4=xgb.predict(x_test_scaled)
accuracy_score(y_test,y_pred4)*100
```
